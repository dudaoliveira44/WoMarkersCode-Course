```python
#1
df_saude_sono.rename(columns = {'ID':'Identificador','Pressão sanguíneaaaa':'Pressão sanguínea','Ocupação':'Profissão','Categoria BMI':'Categoria IMC'}, inplace = True)
df_saude_sono.columns
```

```python
#2
df_saude_sono.groupby('Profissão')['Duração do sono'].agg(Média = 'mean', Mediana = 'median', Moda = pd.Series.mode)
```

```python
#3
is_obeso = df_saude_sono['Categoria IMC'] == 'Obesidade'
is_engsoftware = df_saude_sono['Profissão'] == 'Eng. de Software'
total_obesos_engsoft = df_saude_sono[is_obeso & is_engsoftware]['Identificador'].count()
num_obesos_engsoft = (total_obesos_engsoft/df_saude_sono['Identificador'].count())
print(f'A porcentagem de pessoas que são simultâneamente obesos e Engenheiros de Software é de: {num_obesos_engsoft:.2%}')
```

```python
#4
media_sono = df_saude_sono['Duração do sono'].mean()
is_advogado_and_repvendas = df_saude_sono[df_saude_sono['Profissão'].isin(['Advogado(a)','Representante de Vendas'])]['Duração do sono']
total_adv_and_repvendas = is_advogado_and_repvendas.count()
maior_med = (is_advogado_and_repvendas > media_sono).sum()

if ( maior_med /  total_adv_and_repvendas ) >= 0.5:
  print(f'Como {maior_med} de {total_adv_and_repvendas} pessoas entre Advogados e Representantes de vendas possuem uma média de sono MAIOR que a média de {media_sono:.2f} horas, significa que dormir mal não tem relação com uma dessas profissões')
else:
  print(f'Como {maior_med} de {total_adv_and_repvendas} pessoas entre Advogados e Representantes de vendas possuem uma média de sono MENOR que a média de {media_sono:.2f} horas, significa que dormir mal tem relação com uma dessas profissões')
```

```python
#5
media_sono_enf = df_saude_sono[df_saude_sono['Profissão'].isin(['Enfermeiro(a)'])]['Duração do sono'].mean()
media_sono_med = df_saude_sono[df_saude_sono['Profissão'].isin(['Médico(a)'])]['Duração do sono'].mean()

if media_sono_enf > media_sono_med:
  print(f'Como os enfermeiros possuem média de {media_sono_enf:.2f} horas de sono e os médicos possuem {media_sono_med:.2f} horas de sono, os médicos possuem uma média de sono MENOR que os enfermeiros.')
else:
  print(f'Como os enfermeiros possuem média de {media_sono_enf:.2f} horas de sono e os médicos possuem {media_sono_med:.2f} horas de sono, os enfermeiros possuem uma média de sono MENOR que os médicos.')
```

```python
#6
df_subconjunto = df_saude_sono[['Identificador', 'Gênero', 'Idade','Pressão sanguínea','Frequência cardíaca']]
df_subconjunto.head()
```

```python
#7
qtd_prof_menos_freq = df_saude_sono['Profissão'].value_counts().min()
prof_menos_freq = df_saude_sono['Profissão'].value_counts().idxmin()
print(f'A profissão de {prof_menos_freq} é a profissão menos frequente no subconjunto, contabilizada apenas {qtd_prof_menos_freq} vez(es).')
```

```python
#8
'''
PAM = (Pressão Sistólica + 2 x Pressão Diastólica)/3
'''
df_saude_sono[['Pressão sistólica', 'Pressão diastólica']] = df_saude_sono['Pressão sanguínea'].str.split('/', expand=True).astype(int)
df_saude_sono['Pressão Média'] = ( df_saude_sono['Pressão sistólica'] + 2*df_saude_sono['Pressão diastólica'] ) / 3
media_genero = df_saude_sono.groupby('Gênero')['Pressão Média'].mean()
pm_homem = media_genero['Homem']
pm_mulher = media_genero['Mulher']

if pm_homem > pm_mulher:
  print(f'Como os homens possuem uma média de {pm_homem:.2f} de pressão sanguínea média e as mulheres possuem {pm_mulher:.2f} de pressão sanguínea média, os homens possuem uma média de pressão sanguínea MAIOR que as mulheres.')
else:
  print(f'Como os homens possuem uma média de {pm_homem:.2f} de pressão sanguínea média e as mulheres possuem {pm_mulher:.2f} de pressão sanguínea média, os homens possuem uma média de pressão sanguínea MENOR que as mulheres.')
```

```python
#9
moda = df_saude_sono['Duração do sono'].agg(pd.Series.mode)[0]

if moda == 8:
  print(f'É predominante entre as pessoas dormir {moda} horas.')
else:
  print(f'Não é predominante entre as pessoas dormir 8 horas e sim por {moda} horas.')
```

```python
#10
mpassos_acima70 = df_saude_sono[ df_saude_sono['Frequência cardíaca'] > 70 ]['Passos diários'].mean()
mpassos_abaixo70 = df_saude_sono[ df_saude_sono['Frequência cardíaca'] <= 70 ]['Passos diários'].mean()

if mpassos_acima70 > mpassos_abaixo70:
  print(f'Pessoas com frequência cardíaca acima de 70 possuem uma média de passos diários de {mpassos_acima70:.2f}, sendo MAIOR que de pessoas com frequência cardíaca abaixo de 70 que possuem uma média de {mpassos_abaixo70:.2f} passos.')
else:
  print(f'Pessoas com frequência cardíaca acima de 70 possuem uma média de passos diários de {mpassos_acima70:.2f}, sendo MENOR que de pessoas com frequência cardíaca abaixo de 70 que possuem uma média de {mpassos_abaixo70:.2f} passos.')
```
